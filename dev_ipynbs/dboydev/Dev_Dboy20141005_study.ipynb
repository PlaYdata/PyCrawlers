```python
date_pattern = re.compile("[0-9]{2,4}/[0-9]{1,2}/[0-9]{1,2}")
```

```python
match_inds = []
for ind in range(len(lines)):
    date = date_pattern.findall(lines[ind])
    if len(date) > 0:
        match_inds.append(ind)
```

```python
match_inds[1:] + [200]
```

```python
test = [ind for ind in match_inds[1:]]+ [len(lines) - 1]
pairs = zip(match_inds, test)
```

```python
pairs
```

```python
df_list = []
for pair in pairs:
    temp_data_string = u"\n".join(lines[pair[0] + 1:pair[1]])
    temp_string_buffer = StringIO(temp_data_string)
    temp_df = pd.read_csv(temp_string_buffer, header=None)
    df_list.append(temp_df)
```
