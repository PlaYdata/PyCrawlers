http://www.gretai.org.tw/web/stock/aftertrading/DAILY_CLOSE_quotes/stk_quote.php?l=en-us

```python
import pandas as pd
from pandas.io.parsers import StringIO
import numpy as np
import requests
import csv
```

```python
res = requests.get("http://www.gretai.org.tw/web/stock/aftertrading/DAILY_CLOSE_quotes/stk_quote_download.php?l=en-us&d=2014/10/03")
```

```python
res.text[:500]
```

```python
res.encoding = "big5"
```

```python
csv_string = res.text.replace(u'\uff0b',u"+").replace(u'\uff0d',u"-").replace(u'\uff1a',u":").replace(u"\u2019", u"`")
```

```python
csvfile = StringIO(csv_string)
```

```python
rows = list(csv.reader(csvfile))
```

```python
columns_num_set = set(map(lambda xx: len(xx), rows))
```

```python
columns_num_set
```

```python
len17_data = [one_row for one_row in rows if len(one_row) == 17]
```

```python
df = pd.DataFrame(len17_data[1:], columns = len17_data[0])
```

```python
df["Date"] = "2014/10/03"
```

```python
df
```

```python
len1_data = [one_row for one_row in rows if len(one_row) == 1]
```
